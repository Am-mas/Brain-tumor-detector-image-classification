```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Create a Sequential model
model = Sequential()

# Add the first Conv2D layer with 32 filters, a 3x3 kernel, and 'relu' activation
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(512, 512, 3)))

# Add a MaxPooling2D layer to downsample the spatial dimensions
model.add(MaxPooling2D(pool_size=(2, 2)))

# Add the second Conv2D layer with 64 filters, a 3x3 kernel, and 'relu' activation
model.add(Conv2D(64, (3, 3), activation='relu'))

# Add another MaxPooling2D layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Add the third Conv2D layer with 128 filters, a 3x3 kernel, and 'relu' activation
model.add(Conv2D(128, (3, 3), activation='relu'))

# Flatten the output to feed into a fully connected layer
model.add(Flatten())

# Add a fully connected layer with 128 neurons and 'relu' activation
model.add(Dense(128, activation='relu'))

# Add the output layer with the appropriate number of output neurons (e.g., for binary classification)
model.add(Dense(1, activation='sigmoid'))

# Compile the model with a chosen optimizer, loss function, and metric(s)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display the model summary
model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,

)


# Create generators for training and validation
train_generator = datagen.flow_from_directory(
    train_subfolder,
    target_size=(512, 512),
    batch_size=32,
    class_mode='categorical',
    subset='training'  # Use the training subset
)

validation_generator = datagen.flow_from_directory(
    validation_subfolder,
    target_size=(512, 512),
    batch_size=32,
    class_mode='categorical',
    subset='validation'  # Use the validation subset
)

# Train the model with a specified number of epochs
epochs = 3  # Change this to the desired number of epochs

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)
```
